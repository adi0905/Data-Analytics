```python
# Problem 3
# A list contains the average daily temperature(in degree Celsius) of a city over a particular week. Write a Python code to swap the highest and the lowest temperatures

# a) A list containing average daily temperature over a week
# temperatures = [34, 40, 29, 33, 42, 37, 39 ]

# b) The expected output



# c)Store the highest temperature


# d) Index of the element with the highest temperature
# max_temp_index =


# e) Store the lowest temperature
# min_temp =

# f) Index of the element with the lowest temperature
# min_temp_index =
```

```python
# b) The expected output

temperatures = [34, 40, 29, 33, 42, 37, 39]
highest = 0
lowest = 0

for i in range(0,7):
    if temperatures[i] >= temperatures[highest]:
        highest = i
    if temperatures[i] <= temperatures[lowest]:
        lowest = i


print("lowest:", temperatures[lowest], "highest:", temperatures[highest])

# Swap the values of lowest and highest without a separate function
lowest, highest = highest, lowest

print("After swapping - lowest:", temperatures[lowest],"-->index:",lowest, "highest:", temperatures[highest],'-->index:',highest)
```

```python

# Problem 7 - numpy
# Construct a 2-D  10∗10  matrix with the following constraints:
# The element at  [0,0]  should be 1
# The element at  [0,9]  should be 100
# The element at  [9,0]  should be 50
# The interval between elements in rows should be constant
# The interval between elements in columns should be constant
```

```python
!pip install numpy
!pip install pandas
import numpy as np
import pandas as pd
```

```python
start_value = 1
end_value_row0 = 100
end_value_row9 = 50
matrix_size = 10

matrix = np.zeros((matrix_size, matrix_size), dtype=int)

interval_row = (end_value_row0 - start_value) / (matrix_size - 1)
interval_col = (end_value_row9 - start_value) / (matrix_size - 1)

matrix[0, 0] = start_value
for col in range(1, matrix_size):
    matrix[0, col] = matrix[0, col - 1] + interval_row

matrix[0, 9] = end_value_row0
for row in range(1, matrix_size):
    matrix[row, 9] = matrix[row - 1, 9] + interval_col

matrix[9, 0] = end_value_row9

print(matrix)
```
